```python
import re
def get_pfn_site(path, rules):
    if isinstance(rules, dict):
        for rule, pfn in rules.items():
            if m:=re.match(rule, path):
                grs = m.groups()
                for i in range(len(grs)):
                    print(m.groups()[i])
                    pfn = pfn.replace(f"${i+1}", grs[i])
                return pfn
    else: 
        return rules + "/"+ path
    
    
```
